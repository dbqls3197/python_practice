```python
# dropna() : 결측치가 있는 행 또는 열 전체를 삭제
.dropna(axis=0) # 결측치가 있는 행 전체를 삭제 axis=0은 생략 가능하다.(기본값)
.dropna(axis=1) # 결측치가 있는 컬럼 전체를 삭제 
```

```python
df_row = df.dropna(how="any")              # 기본값(how = 'any') : NaN이 하나라도 있는 행 삭제
df_col = df.dropna(axis=1, how="any")      # 기본값(how = 'any') : NaN이 하나라도 있는 열 삭제

df_row = df.dropna(how="all")              # 모든 값이 NaN 인 행만 삭제 
df_col = df.dropna(axis=1, how="all")      # 모든 값이 NaN 인 열만 삭제 

df_row = df.dropna(thresh=3)               # NaN이 특정 개수 N개 이상인 행 삭제
df_col = df.dropna(axis=1, thresh=3)       # NaN이 특정 개수 N개 이상인 열 삭제
```

```python
import pandas as pd
df = pd.DataFrame({
    '제품': ['A', 'A', 'B', 'B', 'C'],
    '가격': [1000, 1200, 800, 850, 900],
    '판매량': [10, 15, 20, 25, 30],
    '평점': [4.5, 4.0, 4.8, 4.2, 4.6]
})
df
```

```python
df.groupby('제품').mean()
```

```python
df.groupby('제품').agg('mean')
```

```python
df.groupby('제품').agg({
        '가격': 'mean',
        '판매량': 'sum',
        '평점':'max'
})
```

```python
df.groupby('제품').agg({
        '가격': ['mean','min','max'],
        '판매량': ['sum','count'],
        '평점':['max','std']
}) 
# std는 표준 편차 단일값만 있다면 NaN으로 표시됨
```

```python
def range_calc(x):
    return x.max() -x.min()

df.groupby('제품').agg({
    '가격':['mean',range_calc],
    '판매량':'sum'
})
```
