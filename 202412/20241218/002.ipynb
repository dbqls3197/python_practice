### .plot()
- 점(마커)과 점 사이에 선을 그어서 선 그래프로 만든다.

```python
plt.plot([10,20,30,40])  # y축으로 표시 (x축: 1,2,3.. 자동)
plt.show()
```

```python
plt.plot([1,2,3,4],[10,20,30,40])  # x축과 y축
plt.show()
```

```python
import numpy as np
x = np.arange(1, 5)
y = np.arange(10, 50,10) # 10, 20, 30, 40
plt.plot(x,y)
plt.show
```

```python
x =np.array([1,2,3,4])
y =np.array([2,4,12,14])
x2 =np.array([3,4,5,6])
y2 =np.array([5,7,15,17])
plt.plot(x,y,'p',label='팬타곤',markerfacecolor ="red", markersize= 10)  # markersize 마커 사이즈 조절
plt.plot(x2,y2,'p',label='다이아몬드',markerfacecolor ="blue", markersize= 7)  # markersize 마커 사이즈 조절
plt.legend() # 범례표시
#plt.axis('off') # 전체 축 숨기기
plt.show
```

```python
x =np.array([1,2,3,4])    # x축의 값
y1 =np.array([2,4,12,14]) # 첫 번째 선에 대한 y축의 값
y2 =np.array([3,5,13,15]) # 두 번쨰 선에 대한 y축의 값

plt.title('그래프 제목',loc="center",fontsize = 24,color='blue', pad=15)
plt.plot(x,y1, label='first line', color='blue')  # color : 선색깔
plt.plot(x,y2, label='second line', color='orange')
plt.legend(title='범례 타이틀', loc='lower right') # 범례
plt.show()
```

```python
plt.plot([1,2,3,4],[28,25,26,21],color='#0cca4a')
#plt.grid(axis= 'both') # axis: 축방향(x,y,bath)
# linestyle(선 스타일)
# linewidth(선너비) : 0 ~ 1
# alpha(투명도)     : 0 ~ 1

plt.grid(color="m",linestyle='-.',linewidth=0.5, alpha=0.5)
plt.show()
```

```python
plt.plot([1,2,3,4],[28,25,26,21])
# 기본 그리드 선 추가
plt.grid(which='major',color="black",linestyle='-.',linewidth=0.75, alpha=1)

# 보조 그리드 선 추가
plt.minorticks_on()
plt.grid(which='major',color="gray",linestyle=':',linewidth=0.5, alpha=0.7)
```

```python
x = [1,2,3,4]
y = [78,75,76,71]

plt.plot(x,y)

plt.xlabel('주간',labelpad=15, fontsize=14, color='r',loc='right', rotation=45)
plt.ylabel('체중',labelpad=20, fontsize=14, color='b',loc='top')
plt.show()
```

```python
x = [1,2,3,4]
y = [78,75,76,71]

plt.plot(x,y)

font = {'family' : 'NanumGothic',
       'color' : 'b',
        'size' : 15,
        'weight' :'bold'}

plt.xlabel('주간', fontdict = font) # 아래와 동일
#plt.xlabel('주간', fontdict={'family' : 'NanumGothic', 'color' : 'b','size' : 15,'weight' :'bild'})

plt.ylabel('체중',fontdict=font)
#plt.show()
```

```python
# 범위 직접 지정

x = [1,2,3,4]
y = [28,25,26,21]

plt.plot(x,y)                # x축의 범위를  0 ~ 5로 설정한다.
plt.xlim(0,5)                # y축의 범위를 10 ~ 30로 설정한다.
plt.ylim(10,30)              # x축의 범위는 왼쪽으로 0에서 오른쪽으로 5 설정한다.
#plt.xlim(left=0,right=5)    # y축의 범위는 아래쪽 20에서 위쪽 30으로 설정한다.
#plt.ylim(bottem, top=30)
plt.axis([0,5,10,30])        #[xmin,xmax,ymin,ymax]
plt.show()
```

```python
import matplotlib.pyplot as plt
data={
    '성명':['홍길동','김철수','이영희','박민수','정하나'],
    '급여':[3000000,4500000,2500000,5000000,3500000]
}
plt.figure(figsize=(10,5))
plt.plot(data['성명'],data['급여'],marker='o',linestyle='--',color='b')
plt.title('선수별 급여')
plt.xlabel('성명')
plt.ylabel('급여(원)')
plt.grid()
plt.show()
```

### bar()
- 데이터 값을 막대 그래프로 표시

```python
import matplotlib.pyplot as plt

x = [1,2,3]
values = [100,700,500]
# x: 막대 그래프의 x축에 표시될 범주형 변수, values: 각 막대 그래프의 크기
# align = 'edge', align='center' # align은 밑에 날짜
plt.bar(x, values,color='skyblue',width=0.6,edgecolor='gray',linewidth=5,align='center')
plt.xticks(x,['2021년','2022년','2023년'])
plt.show()
```

```python
x = [1,2,3,4]
revenue = [20,21,25,25]
profit = [15,18,19,20]

plt.bar(x, revenue,color='blue',width=0.6,align='center', label='매출')
plt.bar(x, profit,color='red',width=0.4,align='edge', label='수익')

plt.xticks(x,['2021년','2022년','2023년','2024년'])
plt.legend()
plt.show()
```
