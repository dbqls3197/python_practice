## 식당주문

```python
'''
- 주문대기열
- 완료된 주문
1) 주문하기
2) 주문완료
'''
from datetime import datetime

class OrderSystem:
    def __init__(self):
        self.orders = [] # 주문 대기열
        self.completed = [] # 완료된 주문내역

    # 주문하기
    def place_order(self, table_num, items):
        # 테이블번호, 음식들, 시간, 상태(웨이팅)
        order = {
            'id' : len(self.orders) + 1,
            'table': table_num,
            'items': items,
            'time' : datetime.now(),
            'status' : 'wating'
        }
        self.orders.append(order)
        
    def complete_order(self):
        if self.orders:
            order = self.orders[0]
            order ['status'] = 'completed'
            self.completed.append(order)
            self.orders.pop(0)
            return order
        return None

r1 = OrderSystem()
r1.place_order(5,["피자","콜라"])
r1.place_order(7,["햄버거","사이다"])
r1.place_order(9,["치킨","맥주"])
print(r1.orders)
print("-" *70)
completed = r1.complete_order()
print(completed)
print("-" *70)
print(r1.orders)
print("-" *70)
completed = r1.complete_order()
print(completed)
print("-" *70)
```

## 도서 대출

```python
'''
- 책 리스트 (책번호, 책정보)
- 대출 리스트(책번호, 대출자, 대출기한)
- 책정보(도서번호, 책명, 대출여부)
1) 입고
2) 대출
3) 반납
'''

from datetime import datetime

class Library:
    def __init__(self):
        self.books = {}
        self.borrowed = {}
        
    # 도서 입고(대출가능)
    def add_book(self, book_code, title):
        self.books[book_code] = {
            'title': title,
            'available': True 
        }
        # 대출 처리
    def borrow_book(self, book_code, user_id):
        if book_code in self.books and self.books[book_code]['available']:
        
            self.books[book_code]['available'] = False
            self.borrowed[book_code] = {
                'user': user_id,
                'due_date' : datetime.now() +timedelta(days=14)
            }
            return True
        return False
    # 반환처리
    def return_book(self, book_code):
        if book_code in self.borrowed:
            self.books[book_code]['available'] = True
            del self.borrowed[book_code]
            return True
        return False

lb1 = Library()    
lb1.add_book("ISBN001","혼자하는 파이썬")
lb1.add_book("ISBN002","둘이하는 파이썬2")

print(lb1.books)
print(lb1.borrowed)

print("-"*100)
# 대출실행
lb1.borrow_book("ISBN001","홍길동")
print(lb1.books)
print(lb1.borrowed)

print("-"*100)
# 반납
lb1.return_book("ISBN001")
print(lb1.books)
print(lb1.borrowed)
```

## 주차장 관리 시스템

```python
'''
- 주차장: 크기(몇 대까지 가능한지)
- 주차위치 : 인덱스, 차량번호
1) 입차
2) 출차
'''

class ParkingLot:
    def __init__(self, capa):
        self.capa =capa
        self.spots = [None] * capa # 주차장 맵
        self.car_location = {} # 주차차량 정보

    def park(self, car_number):
        if len(self.car_location) >= self.capa:
            return False, "주차장이 다 찼습니다."

        for i in range(self.capa):
            if self.spots[i] is None:
                self.spots[i] = car_number
                self.car_location[car_number] = i
                return True, f"{car_number} 차량{i+1}번 위치 주차완료"
                
    def exit_car(self, car_number):
        if car_number in self.car_location:
            spot = self.car_location[car_number]
            self.spots[spot] = None
            del self.car_location[car_number]
            return True, f"출차 완료: {spot+1}번 자리"
            
        return False, "해당 차량을 찾을 수 없습니다."

p1 = ParkingLot(20)
result, message = p1.park('서울1234')
print(message)
print(p1.spots)
print(p1.car_location)

print("-"*100)

result, message = p1.park('경기5678')
print(message)
print(p1.spots)
print(p1.car_location)

print("-"*100)

result, message = p1.exit_car("서울1234")
print(message)
print(p1.spots)
print(p1.car_location)

print("-"*100)

result, message = p1.park('제주9999')
print(message)
print(p1.spots)
print(p1.car_location)

print("-"*100)

        
```
