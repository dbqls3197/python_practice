```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By

def setup_selenium():
    try:
        # Chrome 옵션 설정
        chrome_options = Options()
        chrome_options.add_argument('--headless')  # 헤드리스 모드
        chrome_options.add_argument('--no-sandbox')
        chrome_options.add_argument('--disable-dev-shm-usage')
        
        # WebDriver 설정
        service = Service('chromedriver.exe')  # 크롬드라이버 경로
        driver = webdriver.Chrome(service=service, options=chrome_options)
        
        # 테스트
        driver.get('https://www.python.org')
        print("페이지 제목:", driver.title)
        
        # 동적 요소 테스트
        search_box = driver.find_element(By.NAME, 'q')
        search_box.send_keys('python')
        search_box.submit()
        
        print("Selenium 설정 완료")
        driver.quit()
        
    except Exception as e:
        print("Selenium 설정 실패:", str(e))
setup_selenium()
```

```python
# 1. 티스토리 블로그 크롤링
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
import time

def tistory_blog():
    url = "https://www.tistory.com/?category=living"
   
    # Chrome 옵션 설정
    options = Options()
    options.add_argument('--headless')  # 브라우저가 보이지 않게 실행
    options.add_argument('--no-sandbox')
    options.add_argument('--disable-dev-shm-usage')

    # Chrome 드라이버 경로 설정
    service = Service('chromedriver.exe')  # chromedriver 경로 지정
    driver = webdriver.Chrome(service=service, options=options)
   
    # 웹 페이지 열기
    driver.get(url)
   
    # 페이지 로딩 대기
    time.sleep(5)  # 필요에 따라 대기 시간 조정
   
    posts = []
   
    # 블로그 포스트 정보 가져오기
    items = driver.find_elements(By.CSS_SELECTOR, 'div.item_group')
    for item in items:
        title = item.find_element(By.CSS_SELECTOR, '.tit_cont').text.strip()
        desc = item.find_element(By.CSS_SELECTOR, '.wrap_desc').text.strip()
        author = item.find_element(By.CSS_SELECTOR, '.link_blog .txt_g').text.strip()
       
        posts.append({
            'title': title,
            'description': desc,
            'author': author
        })
   
    # HTML 생성
    html_content = f'''
    <html>
    <head>
        <style>
            .blog-container {{
                max-width: 900px;
                margin: 20px auto;
                font-family: 'Noto Sans KR', sans-serif;
            }}
            .blog-post {{
                background: #fff;
                padding: 20px;
                margin: 15px 0;
                box-shadow: 0 2px 5px rgba(0,0,0,0.1);
                border-radius: 8px;
            }}
            .blog-title {{
                font-size: 20px;
                color: #333;
                margin-bottom: 10px;
            }}
            .blog-desc {{
                color: #666;
                line-height: 1.6;
            }}
            .blog-author {{
                color: #888;
                font-size: 14px;
                margin-top: 10px;
            }}
        </style>
    </head>
    <body>
        <div class="blog-container">
    '''
   
    for post in posts:
        html_content += f'''
            <div class="blog-post">
                <div class="blog-title">{post['title']}</div>
                <div class="blog-desc">{post['description']}</div>
                <div class="blog-author">작성자: {post['author']}</div>
            </div>
        '''
   
    html_content += '''
        </div>
    </body>
    </html>
    '''
   
    # HTML 파일 저장
    with open('tistory_posts.html', 'w', encoding='utf-8') as f:
        f.write(html_content)

    driver.quit()  # 드라이버 종료

# 크롤링 함수 호출
tistory_blog()
```

```python
# 1. 티스토리 블로그 크롤링
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
import time

def brunch():
    url = "https://brunch.co.kr/search?q=일상"
   
    # Chrome 옵션 설정
    options = Options()
    options.add_argument('--headless')  # 브라우저가 보이지 않게 실행
    options.add_argument('--no-sandbox')
    options.add_argument('--disable-dev-shm-usage')

    # Chrome 드라이버 경로 설정
    service = Service('chromedriver.exe')  # chromedriver 경로 지정
    driver = webdriver.Chrome(service=service, options=options)
   
    # 웹 페이지 열기
    driver.get(url)
   
    # 페이지 로딩 대기
    time.sleep(5)  # 필요에 따라 대기 시간 조정
   
    posts = []
   #resultArticle > div > div.result_article > div.wrap_article_list > ul > li:nth-child(2) > a > div.post_title.has_image
    #resultArticle > div > div.result_article > div.wrap_article_list > ul > li:nth-child(2) > a > div.post_title.has_image > span
    # 블로그 포스트 정보 가져오기
    items = driver.find_elements(By.CSS_SELECTOR, 'div.post_title.has_image')
    
    for item in items:
        title = item.find_element(By.CSS_SELECTOR, '.tit_subject').text.strip()
        desc = item.find_element(By.CSS_SELECTOR, '.article_content').text.strip()
        author = item.find_elements(By.CSS_SELECTOR, '.mobile_d_n.post_append span')[-1].text.strip()
       
        posts.append({
            'title': title,
            'description': desc,
            'author':author
        })
    
   
    # HTML 생성
    html_content = f'''
    <html>
    <head>
        <style>
            .blog-container {{
                max-width: 900px;
                margin: 20px auto;
                font-family: 'Noto Sans KR', sans-serif;
            }}
            .blog-post {{
                background: #fff;
                padding: 20px;
                margin: 15px 0;
                box-shadow: 0 2px 5px rgba(0,0,0,0.1);
                border-radius: 8px;
            }}
            .blog-title {{
                font-size: 20px;
                color: #333;
                margin-bottom: 10px;
            }}
            .blog-desc {{
                color: #666;
                line-height: 1.6;
            }}
            .blog-author {{
                color: #888;
                font-size: 14px;
                margin-top: 10px;
            }}
        </style>
    </head>
    <body>
        <div class="blog-container">
    '''
   
    for post in posts:
        html_content += f'''
            <div class="blog-post">
                <div class="blog-title">{post['title']}</div>
                <div class="blog-desc">{post['description']}</div>
                <div class="blog-author">작성자: {post['author']}</div>
            </div>
        '''
   
    html_content += '''
        </div>
    </body>
    </html>
    '''
   
    # HTML 파일 저장
    with open('brunch.html', 'w', encoding='utf-8') as f:
        f.write(html_content)

    driver.quit()  # 드라이버 종료

# 크롤링 함수 호출
brunch()
```
