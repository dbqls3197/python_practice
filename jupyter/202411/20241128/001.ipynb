```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
import time

def brunch():
    url = "https://brunch.co.kr/search?q=일상"
   
    # Chrome 옵션 설정
    options = Options()
    options.add_argument('--headless')  # 브라우저가 보이지 않게 실행
    options.add_argument('--no-sandbox')
    options.add_argument('--disable-dev-shm-usage')

    # Chrome 드라이버 경로 설정
    service = Service('chromedriver.exe')  # chromedriver 경로 지정
    #driver = webdriver.Chrome(service=service, options=options)
    driver = webdriver.Chrome(options=options) # 셀레니움 4.x 이상
   
    # 웹 페이지 열기
    driver.get(url)
   
    # 페이지 로딩 대기
    time.sleep(5)  # 필요에 따라 대기 시간 조정
   
    posts = []
    # 블로그 포스트 정보 가져오기
    items = driver.find_elements(By.CSS_SELECTOR, 'div.post_title.has_image')
    
    for item in items:
        title = item.find_element(By.CSS_SELECTOR, '.tit_subject').text.strip()
        desc = item.find_element(By.CSS_SELECTOR, '.article_content').text.strip()
        author = item.find_elements(By.CSS_SELECTOR, '.mobile_d_n.post_append span')[-1].text.strip()
       
        posts.append({
            'title': title,
            'description': desc,
            'author':author
        })
    
   
    # HTML 생성
    html_content = f'''
    <html>
    <head>
        <style>
            .blog-container {{
                max-width: 900px;
                margin: 20px auto;
                font-family: 'Noto Sans KR', sans-serif;
            }}
            .blog-post {{
                background: #fff;
                padding: 20px;
                margin: 15px 0;
                box-shadow: 0 2px 5px rgba(0,0,0,0.1);
                border-radius: 8px;
            }}
            .blog-title {{
                font-size: 20px;
                color: #333;
                margin-bottom: 10px;
            }}
            .blog-desc {{
                color: #666;
                line-height: 1.6;
            }}
            .blog-author {{
                color: #888;
                font-size: 14px;
                margin-top: 10px;
            }}
        </style>
    </head>
    <body>
        <div class="blog-container">
    '''
   
    for post in posts:
        html_content += f'''
            <div class="blog-post">
                <div class="blog-title">{post['title']}</div>
                <div class="blog-desc">{post['description']}</div>
                <div class="blog-author">작성자: {post['author']}</div>
            </div>
        '''
   
    html_content += '''
        </div>
    </body>
    </html>
    '''
   
    # HTML 파일 저장
    with open('brunch.html', 'w', encoding='utf-8') as f:
        f.write(html_content)

    driver.quit()  # 드라이버 종료

# 크롤링 함수 호출
brunch()
```

```python
from selenium import webdriver
#from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
import requests
import json
import time
import os
import httpx
import ssl

ssl_context = ssl.create_default_context()
ssl_context.options |= ssl.OP_LEGACY_SERVER_CONNECT

def crawl_yonhap_hotnews():
    
    url = "https://www.yna.co.kr/theme/hotnews-history"

    if not os.path.exists('images'):
        os.makedirs('images')
    
    options = Options()
    options.add_argument('--headless')

    driver = webdriver.Chrome(options=options)
    driver.get(url)
    
    time.sleep(5)
    
    articles = []
    html_content = ['<!DOCTYPE html>',
                   '<html>',
                   '<head>',
                   '<meta charset="utf-8">',
                   '<title>연합뉴스 핫뉴스</title>',
                   '<style>',
                   '.article { margin: 20px; padding: 15px; border: 1px solid #ddd; }',
                   '.article img { max-width: 100%; height: auto; }',
                   '</style>',
                   '</head>',
                   '<body>']

    items = driver.find_elements(By.CSS_SELECTOR, 'div.item-box01')

    nn = 1
    for item in items:
        title = item.find_element(By.CSS_SELECTOR, '.tit-news').text.strip()
        desc = item.find_element(By.CSS_SELECTOR, 'p.lead').text.strip()
        timestamp = item.find_element(By.CSS_SELECTOR, '.txt-time').text.strip()

        img_element = item.find_element(By.CSS_SELECTOR, 'img')
        img_url = img_element.get_attribute('src')

        if img_url:
            response = httpx.get(img_url, verify=ssl_context)
            #response = requests.get(img_url)
            filename = f'images/article_{nn}.jpg'
            if response.status_code == 200:
                with open(filename, 'wb') as f:
                    f.write(response.content)
        else:
            filename = None
        
        articles.append({
            'title': title,
            'description': desc,
            'timestamp': timestamp
        })

        html_content.append(f'<div class="article">')
        html_content.append(f'<h2>{title}</h2>')
        if filename:
            html_content.append(f'<img src="{filename}" alt="{title}">')
        html_content.append(f'<p>{desc}</p>')
        html_content.append(f'<small>{timestamp}</small>')
        html_content.append('</div>')
        nn += 1

    html_content.append('</body></html>')
    
    with open('yonhap_hotnews2.json', 'w', encoding='utf-8') as f:
        json.dump(articles, f, ensure_ascii=False, indent=4)

    with open('yonhap_hotnews2.html', 'w', encoding='utf-8') as f:
        f.write('\n'.join(html_content))

crawl_yonhap_hotnews()
```

```python
import requests
from bs4 import BeautifulSoup
import json
from urllib.parse import urljoin
import time
from datetime import datetime

# 헤더 설정
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}

def naver_blog_search(keyword):
    url = f"https://search.naver.com/search.naver?ssc=tab.blog.all&sm=tab_jum&query={keyword}"
    res = requests.get(url, headers=headers)
    soup = BeautifulSoup(res.text, 'html.parser')

    items = soup.select('li.bx .view_wrap')

    posts = []
    
    for item in items:
        title = item.select_one('.title_area a').text
        link = item.select_one('.title_area a')['href']
        summary = item.select_one('.dsc_area a').text
        
        posts.append({
            'title': title,
            'link': link,
            'summary': summary
        })

    with open(f'naver_blog_{keyword}.json', 'w', encoding='utf-8') as f:
        json.dump(posts, f, ensure_ascii=False, indent=2)


naver_blog_search('빅데이터')
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import json
import time
from datetime import datetime
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service

def naver_blog_search(keyword):
    options = webdriver.ChromeOptions()
    options.add_argument('--headless')
    options.add_argument('--no-sandbox')
    options.add_argument('--disable-dev-shm-usage')
    options.add_argument('--user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36')

    service = Service(ChromeDriverManager().install())
    driver = webdriver.Chrome(service=service, options=options)

    url = f"https://search.naver.com/search.naver?ssc=tab.blog.all&sm=tab_jum&query={keyword}"
    driver.get(url)

    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)

    posts = []

    items = driver.find_elements(By.CSS_SELECTOR, 'li.bx .view_wrap')

    for item in items:
        title = item.find_element(By.CSS_SELECTOR, '.title_area a').text
        link = item.find_element(By.CSS_SELECTOR, '.title_area a').get_attribute('href')
        summary = item.find_element(By.CSS_SELECTOR, '.dsc_area a').text

        posts.append({
            'title': title,
            'link': link, 
            'summary': summary
        })

    with open(f'naver_blog2_{keyword}.json', 'w', encoding='utf-8') as f:
        json.dump(posts, f, ensure_ascii=False, indent=2)
        
    driver.quit()
    
naver_blog_search('빅데이터')
```
