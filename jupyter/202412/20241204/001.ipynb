```python
## ActionChains (드래그앤드랍) 
from selenium import webdriver
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.by import By

driver = webdriver.Chrome()
driver.get('http://lecture.pul.kr/ex5.html?act=dragdrop')

source = driver.find_element(By.ID,'draggable')
target = driver.find_element(By.ID,'droppable')

actions = ActionChains(driver)

actions.drag_and_drop(source,target).perform()
```

```python
## ActionChains (정렬)
from selenium import webdriver
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.by import By

driver = webdriver.Chrome()
driver.get('http://lecture.pul.kr/ex5.html?act=sortable')

item1 = driver.find_element(By.ID,'item1')
item2 = driver.find_element(By.ID,'item2')

# 인스턴스 (객체) 생성
actions = ActionChains(driver)

actions.click_and_hold(item1)\
    .move_to_element(item2)\
    .move_by_offset(0,5)\
    .pause(0.5)\
    .release()\
    .perform()
```
