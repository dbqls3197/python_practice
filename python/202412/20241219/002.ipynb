```python
import matplotlib.pyplot as plt
ratio = [34,32,16,18]

labels = ['사과','바나나','메론','포도']
colors = ['gold','silver','lightgray','whitesmoke']
explode = [0.1,0,0,0]
# explode : 차트의 중심에서 벗어나 돌출되도록 만들떄
# startangle : 3시방향(0도), 12시 (90도)
plt.pie(ratio,labels=labels ,autopct='%.1f%%', colors=colors,shadow=True,startangle=-30,explode=explode)
plt.show()
```

```python
colors = ['gold','silver','lightgray','whitesmoke']
explode = [0.1,0,0,0]
wedeprops = {'width':0.7, 'edgecolor':'k','linewidth':1}

# counterclock =False : 시계방향
# counterclock =True : 반시계방향(기본값)
# wedgeprops: 파이 차트 영역의 스타일을 설정
# width: 원크기 (0 ~ 1)
plt.pie(ratio, labels=labels ,autopct='%.1f%%', colors=colors,shadow=True, startangle=-30,explode=explode,wedgeprops=wedgeprops)
plt.axis('equal') # 파이 차트를 원형으로 유지
plt.show()
```
