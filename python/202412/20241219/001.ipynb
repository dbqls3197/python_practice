## 수평막대

```python
x = [1,2,3]
values = [100, 700, 500]
plt.barh(x,values)
plt.show()
```

```python
# height: 상하폭 비율(0~1), edgecolor: 테두리 색상, linewidth : 테두리 두께
plt.barh(x,values,color='b', height=0.6,edgecolor='black',linewidth=2) 
plt.yticks(x,['2021년','2022년','2023년'])
plt.show()
```

```python
x = [1,2,3]
values = [100, 700, 500]
colors =['r','b','g'] # 개별바 컬러 저장
plt.barh(x,values,color=colors, height=0.6,edgecolor='black',linewidth=2) 
plt.yticks(x,['2021년','2022년','2023년'])
plt.show()
```

```python
x = [1,2,3,4]
revenue = [20,21,22,25]
profit = [15,18,19,20]
plt.barh(x,revenue,color='b', height=0.4,align='center', label='매출')
plt.barh(x,profit,color='red', height=0.4,align='edge', label='수익')
plt.legend(loc='lower right') # 범례
plt.yticks(x,['2021년','2022년','2023년','2024년'])
plt.show()
```

### 산점도(스캐터) 그래프

```python
x = [5, 5, 6, 2, 5, 4]
y = [2, 4, 2, 5, 6, 6]

plt.scatter(x,y)
plt.show()
```

```python
x = [5, 5, 6, 2, 5, 4]
y = [2, 4, 2, 5, 6, 6]

plt.scatter(x,y,s = 100, marker='^',color='blue',alpha=0.5,edgecolors='black',linewidths=0.5)
plt.xlim(1,7)
plt.ylim(1,7)
plt.show()
```

```python
x = [5, 5, 6, 2, 5, 4]
y = [2, 4, 2, 5, 6, 6]
sizes = [100,200,300,400,500,600]
colors = [1,2,3,4,5,6] # 1: 가장 어두운 색깔, 6: 가장 밝은 색깔

plt.scatter(x,y, s=sizes, c=colors, alpha=0.6, cmap='plasma')
plt.xlim(1,7)
plt.ylim(1,7)

plt.title("예상 서비스와 서비스 만족도")
plt.xlabel("에상 서비스")
plt.ylabel("서비스 만족도")

plt.colorbar() # 컬러맵에 대한 컬러바를 추가
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 색상 맵 가져오기
cmap = plt.get_cmap('viridis')

# 0에서 1 사이의 값에 대한 색상 샘플 생성
colors = [cmap(i) for i in np.linspace(0, 1, 20)]  # 20개의 색상 샘플
print(colors)

# 색상 시각화
plt.imshow([colors], aspect='auto')
plt.axis('off')
plt.title('Viridis Color Map Sample')
plt.show()
```

```python
'''
[ha (horizontal alignment): 수평 정렬 방식]
- 'center': 중앙 정렬
- 'left': 왼쪽 정렬
- 'right': 오른쪽 정렬

[va (vertical alignment): 수직 정렬 방식]
- 'center': 중앙 정렬
- 'top': 위쪽 정렬
- 'bottom': 아래쪽 정렬
- 'baseline': 기준선 정렬
'''
```

```python
data = {
    '성명': ['홍길동', '김철수', '이영희', '박민수', '정하나'],
    '나이': [28, 35, 29, 42, 31],
    '급여': [3000000, 4500000, 2500000, 5000000, 3500000]   
}
plt.figure(figsize=(10,5))
plt.scatter(data['나이'],data['급여'],s=200,c='red')

for i in range(len(data['성명'])):
    plt.text(data['나이'][i],data['급여'][i]+90000,data['성명'][i],fontsize=10,ha='center',va='bottom')
plt.title('나이에 따른 급여 분포')
plt.xlabel('나이')
plt.ylabel('급여(원)')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.random.rand(10)    # np.random.rand 0 ~ 1 사이의 난수
y = np.random.rand(10)

plt.scatter(x,y,marker='o',color='b',label='원형 마커')
plt.scatter(x+0.1,y+0.1,marker='s',color='red',label='정사각형 마커')
plt.scatter(x+0.2,y+0.2,marker='^',color='green',label='삼각형 마커')
plt.scatter(x+0.3,y+0.3,marker='x',color='purple',label='엑스 마커')


plt.title('다양한 마커 모양')
plt.xlabel('X축')
plt.ylabel('Y축')
plt.legend()
plt.grid()
plt.show()
```
